```python
# PHẦN 1: XỬ LÝ VÀ CHUẨN BỊ DỮ LIỆU

# I. KHỞI TẠO VÀ IMPORT THƯ VIỆN
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# II. XỬ LÝ DỮ LIỆU BAN ĐẦU

# 1. Đọc dữ liệu
df= pd.read_csv('/content/data/dataset.csv',dtype={'explicit': bool})
print("Phân tích ban đầu:")
print("-" * 50)
print(f"Số lượng bài hát ban đầu là: {len(df)}")
#2.Kiểm tra dữ liệu bị thiếu
missing_values= df.isnull().sum()
print(f"Số lượng dòng bị thiếu giá trị là: {len(missing_values)}")
#3.Kiểm tra dữ liệu bị trùng lặp
duplicates = df[df.duplicated(subset=['track_id'])]
print(f"Số lượng bài hát bị trùng lặp: {len(duplicates)}")
# 4. Tạo bản sao sạch của dữ liệu
df_clean= df.dropna()
df_clean=df.drop_duplicates(subset=["track_id"])
```

```python
# 5. Xử lý dữ liệu popularity = 0
zero_pop = df_clean[df_clean['popularity'] == 0]
print(f"Số bài hát có giá trị popularity=0: {len(zero_pop)}")
df_clean=df_clean[df_clean['popularity'] > 0]
```

```python
# III. CHUYỂN ĐỔI VÀ CHUẨN HÓA DỮ LIỆU
#1. Chuyển đổi kiểu dữ liệu
df_clean['instrumentalness'] = df_clean['instrumentalness'].astype(float)
df_clean['duration_min'] = df_clean['duration_ms'] / 60000
df_clean['is_explicit'] = df_clean['explicit'].astype(int)
# 2. Chọn các thuộc tính âm nhạc cần phân tích
audio_features = [
    'danceability',    # Khả năng tạo cảm giác nhảy múa
    'energy',          # Mức độ năng lượng của bài hát
    'loudness',        # Độ lớn của âm thanh
    'speechiness',     # Tỷ lệ từ ngữ trong bài hát
    'acousticness',    # Mức độ acoustic
    'instrumentalness',# Mức độ sử dụng nhạc cụ trong bài hát
    'liveness',        # Xác suất bài hát được trình diễn trực tiếp
    'valence',         # Mức độ tích cực về mặt âm nhạc
    'tempo'            # Tốc độ của bài hát
]
# 3.Chuẩn hóa các tính chất
scaler = StandardScaler()
df_clean[audio_features] = scaler.fit_transform(df_clean[audio_features])
```

```python
# IV. PHÂN TÍCH THỐNG KÊ CƠ BẢN
print("Phân tích sau cùng:")
print("-" * 50)
print(f"Số bài hát sau khi lọc: {len(df_clean)}")
print(f"Tổng số nghệ sĩ trong dataset: {df_clean['artists'].nunique()}")
print(f"Tổng số album trong dataset: {df_clean['album_name'].nunique()}")
print(f"Thời gian trung bình của bài nhạc trong dataset: {df_clean['duration_min'].mean():.2f} phút")
print(f"Trung bình số điểm popularity: {df_clean['popularity'].mean():.2f}")
```

```python
# PHẦN 2: PHÂN TÍCH VÀ TRỰC QUAN HÓA
# I. PHÂN TÍCH TÍNH NĂNG ÂM NHẠC

# 1. Thống kê tính năng âm nhạc
print("\nThống kê các tính năng âm nhạc sau khi Chuẩn hóa:")
print("-" * 50)
print(df_clean[audio_features].describe().round(3))

# 2. Phân tích popularity
print("\nPhân tích theo Popularity:")
print("-" * 50)
print("\nThống kê liên quan đến Popularity:")
print(df_clean['popularity'].describe())
# 3. Phân loại popularity
df_clean['popularity_range'] = pd.cut(df_clean['popularity'],
                                    bins=[0, 20, 40, 60, 80, 100],
                                    labels=['Very Low', 'Low', 'Normal', 'High', 'Very High'])
popularity_dist = df_clean['popularity_range'].value_counts()
print("\nPhân bố Popularity:")
print(popularity_dist)

#   II. PHÂN TÍCH TƯƠNG QUAN
# Tính toán sự tương quan của các tính năng
correlation_matrix = df_clean[audio_features + ['popularity']].corr()
print("\nTương quan với Popularity:")
print(correlation_matrix['popularity'].sort_values(ascending=False))

# III. PHÂN TÍCH THEO NGHỆ SĨ ALBUM VÀ THỜI GIAN
print("\nPhân tích theo nghệ sĩ và album:")
print("-" * 50)

# 1. Top nghệ sĩ
artist_counts = df_clean['artists'].value_counts().head(10)
print("\nTop 10 nghệ sĩ có nhiều bài hát nhất:")
print(artist_counts)

# 2. Top album
album_popularity = df_clean.groupby('album_name')['popularity'].mean().sort_values(ascending=False).head(10)
print("\nTop 10 album có popularity cao nhất:")
print(album_popularity)

# 3. Phân tích thời lượng
print("\nPhân tích thời lượng bài hát:")
print("-" * 50)
print("\nThống kê thời lượng (phút):")
print(df_clean['duration_min'].describe())

# IV. PHÂN TÍCH NỘI DUNG NHẠY CẢM
print("\nPhân tích nội dung nhạy cảm:")
print("-" * 50)

explicit_counts = df_clean['explicit'].value_counts()
explicit_percentage = (explicit_counts / len(df_clean)) * 100

print("\nTỷ lệ bài hát có nội dung nhạy cảm:")
print(f"Có nội dung nhạy cảm: {explicit_percentage[True]:.1f}%")
print(f"Không có nội dung nhạy cảm: {explicit_percentage[False]:.1f}%")


# X. TRỰC QUAN HÓA DỮ LIỆU
plt.figure(figsize=(15, 10))

# 1. Biểu đồ phân bố popularity
plt.subplot(2, 2, 1)
sns.histplot(data=df_clean, x='popularity', bins=20)
plt.title('Phan bo diem Popularity')

# 2. Biểu đồ phân bố tính năng âm thanh
plt.subplot(2, 2, 2)
sns.boxplot(data=df_clean[audio_features])
plt.xticks(rotation=45)
plt.title('Phan bo cac tinh nang thanh')

# 3. Ma trận tương quan
plt.subplot(2, 2, 3)
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Ma tran tuong quan')

# 4. Biểu đồ phân tán
plt.subplot(2, 2, 4)
sns.scatterplot(data=df_clean, x='energy', y='loudness', hue='popularity_range')
plt.title('Energy vs Loudness')

plt.tight_layout()
plt.show()
```

```python
# PHẦN 3:XỬ LÝ PHÂN CỤM
#1. Thêm thư viện
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from yellowbrick.cluster import KElbowVisualizer

# I. PHÂN CỤM DỮ LIỆU
# 1.Chuẩn bị dữ liệu cho phân cụm
features_to_cluster = ['danceability', 'energy', 'valence', 'tempo', 'loudness', 'acousticness']

X = df_clean[features_to_cluster]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 2.Tìm số cụm tối ưu
model = KMeans()
visualizer = KElbowVisualizer(model, k=(2,10))
visualizer.fit(X_scaled)
optimal_k = visualizer.elbow_value_

#3. Thực hiện phân cụm theo Kmeans
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
df_clean['cluster'] = kmeans.fit_predict(X_scaled)

#4.Phân tích các cụm
cluster_stats = df_clean.groupby('cluster')[features_to_cluster].mean()
print("\nCác tâm cụm là:")
print(cluster_stats)

silhouette_avg = silhouette_score(X_scaled, df_clean['cluster'])
print(f"\nSilhouette Score: {silhouette_avg:.3f}")

# XI. TRỰC QUAN HÓA PHÂN CỤM

# 1. Biểu đồ phân cụm
plt.figure(figsize=(10, 6))
plt.scatter(df_clean['energy'], df_clean['valence'], c=df_clean['cluster'], cmap='viridis')
plt.xlabel('Energy')
plt.ylabel('Valence')
plt.title('Phan cum theo Energy và Valence')
plt.colorbar(label='Cluster')
plt.show()

# Lưu mô hình phân cụm
import joblib
joblib.dump(kmeans, 'kmeans_model.pkl')
joblib.dump(scaler, 'scaler.pkl')
```

```python
#2. Phân tích về các cụm
def analyze_clusters(df, features):
    for cluster in df['cluster'].unique():
        cluster_data = df[df['cluster'] == cluster]
        print(f"\nCụm {cluster}:")
        print(f"Số bài hát: {len(cluster_data)}")
        print(f"Điểm popularity trung bình: {cluster_data['popularity'].mean():.2f}")

        # Đặc trưng âm nhạc trung bình
        print("Đặc trưng âm nhạc trung bình:")
        print(cluster_data[features].mean())

        # Top songs
        print("\nTop bài hát:")
        print(cluster_data.nlargest(3, 'popularity')[['track_name', 'artists', 'popularity']])

analyze_clusters(df_clean, features_to_cluster)
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
for i, feature in enumerate(features_to_cluster):
    sns.boxplot(data=df_clean, x='cluster', y=feature, ax=axes[i//3, i%3])
plt.tight_layout()
plt.show()

```

```python
# PHẦN 4: HỆ THỐNG GỢI Ý
def recommend_songs(song_id, n_recommendations=5):
    # Lấy đặc trưng của bài hát đầu vào
    song_features = df_clean[df_clean['track_id'] == song_id][features_to_cluster]
    # Dự đoán cluster của bài hát
    song_cluster = kmeans.predict(scaler.transform(song_features))[0]
    # Lọc các bài hát cùng cluster
    cluster_songs = df_clean[df_clean['cluster'] == song_cluster]
    # Tính độ tương tự
    similarities = cluster_songs.apply(lambda x: np.linalg.norm(
        song_features.values - x[features_to_cluster].values), axis=1)
    #Lấy n bài hát tương tự nhất
    recommendations = cluster_songs.loc[similarities.nsmallest(n_recommendations+1).index[1:]]
    return recommendations[['track_name', 'artists', 'popularity']]
```
